# AFT - Parameter Search Example Notebook

This notebook illustrates an example of searching for an optimal length scale difference between two samples

### Import packages

```python
import skimage.io as io                 # For reading in images
import glob as glob                     # For creating lists of filenames
import AFT_tools as AFT                 # AFT (Alignment by Fourier Transform) functions
import numpy as np                      # For math
import matplotlib.pyplot as plt         # For plotting
import os                               # For saving to disk

%matplotlib notebook
%matplotlib widget
```

### Read in images and define parameters

<div style="background-color:#abd9e9; border-radius: 5px; padding: 10pt"><strong>User to-do:</strong> in the next cell, update paths to images from two samples to be compared and path to folder where to save output</div>

```python
path_to_sample1 = '../../../data/AFT_test_data/images/fig4/anisotropic/'
path_to_sample2 = '../../../data/AFT_test_data/images/fig4/isotropic/'
path_output =  '../../../data/AFT_test_data/output/'
```

```python
# read in images
im1_list = sorted(glob.glob(path_to_sample1+'*.tif'))
im2_list = sorted(glob.glob(path_to_sample2+'*.tif'))
```

<div style="background-color:#abd9e9; border-radius: 5px; padding: 10pt"><strong>User to-do:</strong> select parameters for the analysis in the next cell</div>

```python
# AFT parameters

#### required parameters ####
min_win_size = 25
win_size_interval = 25
overlap = 0.5
```

### Perform parameter search

```python
# perform parameter search: display order parameter for all parameter combinations (sample 1)
Order_dataframe1, window_neighborhood1 = AFT.parameter_search(im1_list, min_win_size, win_size_interval, overlap, plot_figure=True)
```

```python
# perform parameter search: display order parameter for all parameter combinations (sample 2)
Order_dataframe2, window_neighborhood2 = AFT.parameter_search(im2_list, min_win_size, win_size_interval, overlap, plot_figure=True)
```

### Parameter comparison

Calculate the difference between order parameters for the two data sets and calculate the p-value for comparisons between window/neighborhood sizes for each data set.

```python
order_diff, p_median, win_size_list, neighborhood_list = AFT.parameter_comparison(Order_dataframe1, window_neighborhood1,
                                                                                  Order_dataframe2, window_neighborhood2,
                                                                                  save_figures=True, save_path=path_output)
```
