# AFT - Single Image Example Notebook

This notebook illustrates an example of measuring alignment in a single image

### Import packages

```python
import skimage.io as io                           # For reading in images
import AFT_tools as AFT                 # AFT (Alignment by Fourier Transform) functions
import matplotlib.pyplot as plt                   # For plotting
import os
%matplotlib notebook
%matplotlib widget
```

### Read in images and define parameters

<div style="background-color:#abd9e9; border-radius: 5px; padding: 10pt"><strong>User to-do:</strong> in the next cell, update path to single image to be tested and path to folder where to save output</div>

```python
path_to_image = '../../../data/AFT_test_data/images/fig4/anisotropic/01actin_aniso.tif'
path_output =  '../../../data/AFT_test_data/output/'
```

```python
# read image
im = io.imread(path_to_image)
```

<div style="background-color:#abd9e9; border-radius: 5px; padding: 10pt"><strong>User to-do:</strong> select parameters for the analysis in the next cell</div>

```python
# AFT parameters

#### required parameters ####
window_size = 65
overlap = 0.5
neighborhood_radius = 5

#### optional parameters ####
# intensity_threshold = 20
# eccentricity_threshold = 0.4
# im_mask = io.imread('mask').astype('bool')
```

### Calculate the local orientation

```python
x, y, u, v, im_theta, im_eccentricity = AFT.image_local_order(im, window_size, overlap, save_path=path_output, 
                                                             plot_overlay=True, plot_angles=True, plot_eccentricity=True,
                                                             save_figures=True)
```

### Calculate the order parameter

```python
im_order_parameter = AFT.calculate_order_parameter(im_theta, neighborhood_radius)
print('The order parameter is {0}'.format(im_order_parameter))
```
